```python
# title을 factor로 형변환
total['title'] = total['title'].astype('category')
print(total['title'].dtype)
```

```python
groups = total.groupby('title')
```

```python
result = dict(list(groups))
```

```python
music_title = total['title'].str.replace('!@#$%^&*\(\)_\{\}\[\]/\:?<>|;','')
music_title
```

```python
error_list = []
for title in result:
    try:
        title_df = result[title]
        title_df.to_csv('C:/Users/whgud/TIL/team project/music/'+title+'_df.csv',index=False, encoding="utf-8-sig")
    except : 
        error_list.append(title)
        print(title , "오류로 저장 실패")
```
