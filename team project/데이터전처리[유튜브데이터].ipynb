```python
# 댓글사용중지를 결측값으로 대체
youtube['댓글 수'] = youtube['댓글 수'].replace('댓글사용중지', np.NaN) 
```

```python
# 결측값으로 대체되었는지 확인
youtube.head()
```

```python
# 조회수에서 필요없는 문자들 제거
youtube['조회수']=youtube['조회수'].str.replace(',','')
youtube['조회수']=youtube['조회수'].str.replace('조회수','')
youtube['조회수']=youtube['조회수'].str.replace('회','')
```

```python
# 조회수에서 필요없는 문자들 제거 확인
youtube.head()
```

```python
# 좋아요 수에서 필요없는 문자들 제거
youtube['좋아요 수']=youtube['좋아요 수'].str.replace(',','')
youtube['좋아요 수']=youtube['좋아요 수'].str.replace('좋아요 ','')
youtube['좋아요 수']=youtube['좋아요 수'].str.replace('개','')
youtube['좋아요 수']=youtube['좋아요 수'].replace('없음',0)
```

```python
# 좋아요 수에서 필요없는 문자들 제거 확인
youtube.head()
```

```python
# 싫어요 수에서 필요없는 문자들 제거
youtube['싫어요 수']=youtube['싫어요 수'].str.replace(',','')
youtube['싫어요 수']=youtube['싫어요 수'].str.replace('싫어요 ','')
youtube['싫어요 수']=youtube['싫어요 수'].str.replace('개','')
youtube['싫어요 수']=youtube['싫어요 수'].replace('없음',0)
```

```python
# 싫어요 수에서 필요없는 문자들 제거 확인
youtube.head()
```

```python
# 댓글수에서 ,제거
youtube['댓글 수'] = youtube['댓글 수'].str.replace(',','')
```
