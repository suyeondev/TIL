```python
# chromedriver 변수 설정
path = 'C:/Users/whgud/driver/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get('https://www.youtube.com/')
```

```python
# 검색용 url 변수 설정
url = 'https://www.youtube.com/results?search_query='
```

```python
# 값 받아올 리스트 변수설정
title_list = mcow['title']
artist_list = mcow['artist']

mv_name = []
uploader_name=[]
view_cnt = []
upload_date=[]
like_cnt = []
dislike_cnt=[]
comment_cnt=[]
comment_stop=[]
```

```python
# 시간 표시
now = datetime.now() 
formattedDate = now.strftime("%Y%m%d")

# 시작시간
print('START DATE : ', now)

# 크롤링 반복문
for i in range(len(mcow)) :
    driver.get(url+mcow['artist'][i]+ ' '+ mcow['title'][i] + " MV")
    time.sleep(1)
    
 
    try:
        mv_url = driver.find_element_by_xpath('//*[@id="video-title"]').get_attribute('href')
        driver.get(mv_url) 
        time.sleep(1)

        title = driver.find_element_by_xpath('//*[@id="container"]/h1/yt-formatted-string').text
        mv_name.append(title)

        view= driver.find_element_by_xpath('//*[@id="count"]/ytd-video-view-count-renderer/span[1]').text
        view_cnt.append(view)

        date = driver.find_element_by_xpath('//*[@id="info-strings"]/yt-formatted-string').text
        upload_date.append(date)
        time.sleep(0.5)
        
        like=driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[6]/div[2]/ytd-video-primary-info-renderer/div/div/div[3]/div/ytd-menu-renderer/div/ytd-toggle-button-renderer[1]/a/yt-formatted-string').get_attribute('aria-label')
        like_cnt.append(like)
        time.sleep(0.5)
        
        dislike = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/div[6]/div[2]/ytd-video-primary-info-renderer/div/div/div[3]/div/ytd-menu-renderer/div/ytd-toggle-button-renderer[2]/a/yt-formatted-string').get_attribute('aria-label')
        dislike_cnt.append(dislike)
        time.sleep(1)
        
  
        driver.execute_script("window.scrollTo(0, 600)")
    
        time.sleep(2)
        
        uploader = driver.find_element_by_xpath('//*[@id="text"]/a').text
        uploader_name.append(uploader)
        time.sleep(0.5)        

        comment = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-watch-flexy/div[5]/div[1]/div/ytd-comments/ytd-item-section-renderer/div[1]/ytd-comments-header-renderer/div[1]/h2/yt-formatted-string/span[2]').text
        comment_cnt.append(comment)
        time.sleep(1)
        print("{}번째 곡 {} -- done".format(i+1, title))
        
    except:
        print("{}번째 곡 {} {} 댓글수 크롤링에러".format(i+1,title,mcow['artist'][i]))
        
        comment_stop.append(mcow['title'][i])
        
    
    
```

```python
# 크롤링해온 데이터로 데이터프레임 생성
youtube_frm = pd.DataFrame({
    '뮤직카우 곡명' :title_list[:941],
    '뮤직카우 가수명':artist_list[:941],
    '뮤직비디오 이름' :mv_name[:941],
    '업로드 채널명' : uploader_name[:941],
    '조회수' : view_cnt[:941],
    '업로드 날짜': upload_date[:941],
    '좋아요 수': like_cnt[:941],
    '싫어요 수': dislike_cnt[:941],
    
})
```
