# DataFrame row index, data조작, index조작

- loc() : 라벨기반의 2차원 인덱싱
- iloc() : 정수값 기반의 2차원 인덱싱

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json

import warnings
warnings.filterwarnings(action = 'ignore') 

from datetime import date, datetime, timedelta
from dateutil.parser import parse

print("numpy version:",np.__version__) 
print("pandas version:",pd.__version__)
```

```python
def seriesInfo(ary):
    print('data \n', ary)
    print('index : ', type(ary.index), ary.index)
    print('value : ', type(ary.values), ary.values)
    
def frmInfo(df):
    display('DataFrame',df)
    print('shape : ', df.shape) # 형태
    print('size : ' , df.size) # 전체 개수
    print('ndim : ' , df.ndim)
    print('index(row) : ', df.index, type(df.index)) # row index
    print('columns : ', df.columns, type(df.columns))
    print('values : \n', df.values, type(df.values))
```

```python
sample_frm = pd.DataFrame(np.arange(10, 22).reshape(3,4),
                         index=['a','b','c'],
                         columns=['col01','col02','col03','col04'])
frmInfo(sample_frm)
```

```python
# loc, iloc
seriesInfo(sample_frm.loc['a'])
print(sample_frm.loc['a'].values, type(sample_frm.loc['a'].values))
```

#### loc함수 : 행 또는 행과 열의 라벨링이 들어가야 함!! 
- 열의 라벨링만 들어가면 오류남! 

#### iloc함수 : 행 또는 행과 열의 정수인덱스가 들어가야 함!!
- 열의 정수인덱스만 들어가면 오류남!

```python
frmInfo(sample_frm.loc[['a','c']])
frmInfo(sample_frm.iloc[[0,2]])
```

```python
score_data = {
    'kor'  : [80,90,70,30],
    'eng'  : [90,70,60,40],
    'math' : [90,60,90,70] 
}
columns = ['kor','eng','math']
index   = ['김지은', '황인범', '김정수', '최호진']

score_frm = pd.DataFrame(score_data, index=index, columns=columns)
frmInfo(score_frm)
```

```python
# 모든 학생의 각 과목 평균 점수를 새로운 열로 추가하라
score_frm['mean']= round((score_frm['kor']+score_frm['eng']+score_frm['math'])/3,2)
frmInfo(score_frm)
# 다른 방법 : score_frm['average']=np.mean(score_frm, axis=1).round(2)
```

```python
# 최호진 학생의 영어 점수를 90점으로 수정하고 평균 점수도 다시 계산하라
score_frm.loc['최호진','eng']=90
score_frm['mean']= round((score_frm['kor']+score_frm['eng']+score_frm['math'])/3,2)
frmInfo(score_frm)
```

```python
# 김지은 학생의 점수를 데이터 프레임으로 만들어라
jieun_df = pd.DataFrame(score_frm.loc['김지은'])
jieun_df 
```

```python
# 김정수 학생의 점수를 시리즈로 출력하라
print(score_frm.loc['김정수'],type(score_frm.loc['김정수']))
```

```python
# 황인범 학생의 국어점수와 수학점수를 100점으로 수정하고 평균 점수도 다시 계산하라
score_frm.loc['황인범',['kor','math']]=100
score_frm['mean']=np.mean(score_frm[['kor','eng','math']], axis=1).round(2)
frmInfo(score_frm)
```

## 정렬
- sort_index : 인덱스를 정렬하여 행 또는 열을 정렬한다.
- sort_values : 열의 값을 기준으로 행을 정렬한다.

```python
sort_frm = pd.DataFrame(np.random.randint(0,10,(6,4)))
sort_frm
```

```python
# pd.date_range(날짜,periods=기간)원하는 기간의 날짜를 만들어주는 함수
sort_frm.index=pd.date_range('20210901', periods=6) 
sort_frm.columns=["A","B","C","D"]
sort_frm
```

```python
# 인덱스와 칼럼 섞기
random_date=np.random.permutation(sort_frm.index) 
random_date
r_sort_frm=sort_frm.reindex(index=random_date,
                           columns=["B","A","D","C"])
r_sort_frm
```

```python
# sort_index(axis=, ascending=) ascending=False: 내림차순
r_sort_frm.sort_index(axis=0,ascending=False) # 행을 내림차순으로 정렬
```

```python
# sort_values(by=, ascending=)
r_sort_frm.sort_values(by=["B","A"],ascending= False)
```

```python
# 타이타닉에서 승객의 나이를 기준으로 내림차순 정렬을 한다면?
datasets.sort_values(by='age',ascending=False)
```

```python
# 타이타닉중 승객에 대해서 성별의 인원수, 나이별 인원수, 선실별 인원수, 사망/생존별 인원수를 구한다면?
# 성별 인원수
print("성별 인원수: \n", datasets['sex'].value_counts().sort_values())
# 나이별 인원수
print("나이별 인원수: \n", datasets['age'].value_counts().sort_values(ascending=True))
# 선실별 인원수
print("선실별 인원수: \n", datasets['pclass'].value_counts().sort_values(ascending=True))
# 사망/생존별 인원수
print("사망/생존별 인원수: \n", datasets['alive'].value_counts().sort_values())
```
