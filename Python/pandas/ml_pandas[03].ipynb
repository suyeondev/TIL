## scraping을 통한 데이터 처리
### BeautifulSoup
### scraping : 정적인 페이지에 접근해서 데이터를 가져오는 것.

```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
response = requests.get('https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds')
print('response : ', response )
```

```python
# response 의 text를 parsing해주는 BeautifulSoup

soup = BeautifulSoup(response.text , 'lxml')
rows = soup.select('div > ul > li')
print('type - ' , type(rows))
print()

etfs = {}
etf_name   = []
etf_market = []
etf_ticker = []
for row in rows : 
    # print(row.text)
    # print("*"* 100)
    try : 
        etf_name = re.findall('^(.*) \(NYSE' , row.text)
        # print(etf_name)
        etf_market = re.findall('\((.*)\|' , row.text)
        # print(etf_market)
        etf_ticker = re.findall('NYSE Arca\|(.*)\)' , row.text)
        # print(etf_ticker)
        if ( (len(etf_name) > 0 ) & (len(etf_market) > 0) & (len(etf_ticker) > 0) ) :
            etfs[ etf_ticker[0] ] = [ etf_market[0] , etf_name[0]]
    except AttributeError as err :
        pass
```

```python
temp_frm=pd.DataFrame(etfs)
temp_frm
```

```python
import seaborn as sns
datasets=sns.load_dataset('titanic')
print(type(datasets))
```

```python
datasets.head()
```

```python
# age를 기준으로 성인과 미성년자를 구별하는 열을 추가한다면?
datasets['adult/child']=datasets.apply(lambda frm : 'adult' if frm['age']>=20 else 'child',axis=1)
datasets
```

```python
# age를 기준으로 20살이 넘으면 성별을 그대로 사용하고, 
# 미만이면 성별에 관계없이 child라고 하는 category 열을 추가하기
datasets['category']=datasets.apply(lambda frm : 'child' if frm['age']<20 else frm['sex'],axis=1)
datasets
```

```python
# embark_town의 문자열 개수를 별도의 컬럼인 embark_len으로 추가하고 싶다면?
datasets['embark_len']=datasets['embark_town'].apply(lambda frm: len(str(frm)))
datasets
```

```python
#15세 이하는 child ,  15~60 까지는 adult , 61 이상은 elderly 로 분류를해서 
#age_division 컬럼에 저장
#lambda 식으로 apply함수를 이용해서 구현한다면?

## 람다 표현식에 조건문 여러개 사용하기

# lambda 매개변수들: 식1 if 조건식1 else 식2 if 조건식2 else 식3

datasets['age_division'] = datasets.apply(lambda frm: 'child' if frm['age']<=15 else 'adult' if frm['age']<=60 else 'elderly', axis=1 )
datasets['age_division'].value_counts()
```

```python
datasets.info()
```

```python
# 승객에 대한 나이(age)와 성별(sex)에 따른 카테고리를 'age/gender' 추가한다
# 조건 1 : 20살이 넘으면 성별을 그대로 사용하고
# 조건 2 : 20살 미만이면 성별에 관계없이 'child' 라고 정의하고 결측값이 있는 경우는 'i don't know'정의한다
condition1=lambda frm: frm['sex'] if frm['age'] >= 20 else 'child'
condition2=lambda frm : "I don't know" if pd.isnull(frm['age']) else condition1(frm)
datasets['age/gender']=datasets.apply(condition2, axis=1)
datasets
```

```python
data = {
    '국어' : [50, 65, 87, 10, 100, 89, 32, 83, 100, 58],
    '영어' : [100, 30, 20, 11, 88, 97, 100, 100, 10, 57],
    '수학' : [100, 59, 88, 29, 99, 29, 22, 99, 29, 50]
}
```

```python
# 실습문제 
# 10명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.
# 1. 학생 이름을 나타내는 열을 포함시키지 않고 데이터프레임 df_score1 을 생성 후, 
df_score1 = pd.DataFrame(data)

# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
df_score1.index=['학생1','학생2','학생3','학생4','학생5','학생6','학생7','학생8','학생9','학생10']
display(df_score1)

# reset_index 명령으로 이 인덱스 열을 명령으로 
# 일반 데이터열로 바꿔 데이터프레임 df_score2을 만든다.
df_score2 = df_score1.reset_index()
display(df_score2)

# 2.학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다
df_score2.set_index('index')
```
