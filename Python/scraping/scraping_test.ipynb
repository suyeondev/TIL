```python
path = 'C:/Users/whgud/driver/chromedriver_win32/chromedriver.exe'
driver = webdriver.Chrome(path)
driver.get('https://www.musicow.com/market?tab=list&sortorder=amt_market_latest_desc')
```

```python
title_list=[]
artist_list=[]
```

```python
for idx in range(1,21): 
    titles = driver.find_element_by_xpath('//*[@id="market_list"]/div[2]/a['+str(idx)+']/div[2]/strong').text
    artists = driver.find_element_by_xpath('//*[@id="market_list"]/div[2]/a['+str(idx)+']/div[2]/span').text
    for title in titles:
        if title[0]=='인':
            #print(titles)
            titles = titles.replace(title[0],'')
    title_list.append(titles)
    artist_list.append(artists)
    print(idx,titles,artists)
    #time.sleep(1)


```

```python
music_frm = pd.DataFrame({
    'title' : title_list,
    'artist': artist_list
})
music_frm
```

```python
# 잘못들어간 곡제목 수정
music_frm.replace('애만들기','애인만들기', inplace=True)
```
